```python
# Method 1: Using map()
string_tuple = ("1", "2", "3", "4", "5")
int_tuple = tuple(map(int, string_tuple))
print(int_tuple)
```

```python
#method 2: Using tuple comprehension
string_tuple = ("10", "20", "30", "40")
int_tuple = tuple(int(x) for x in string_tuple)
print(int_tuple) 
```

```python
#finding size using getsizeof(tuple,set,list)
from sys import getsizeof
set_items=set((7,2,4,"f",3))
print("set_list:",getsizeof(set_items))

tuple_items=set(1,22,3,[3,4],9)#in set only 1 argument can pass
print("tuple_list:",getsizeof(tuple_items))

tuple_items=set({3,[5,'n',7],6,4})#in set only 1 argument can pass
print("tuple_list:",getsizeof(tuple_items))

```

```python
#DICTIONARY
#create an empty dict and add some key value pairs
#method 1:adding elemts by using key
empty_dict={}
print(empty_dict)
empty_dict['key']='value'
empty_dict['name']='priiii'

print(empty_dict)
```

```python
#DICTIONARY
#create an empty dict and add some key value pairs
#method 2:adding elemts by update()
empty_dict={}
print(empty_dict)
empty_dict.update({'key':'value','name':'Leo'})
print(empty_dict)
```

```python
#DICTIONARY
#create an empty dict and add some key value pairs
#method 1:adding elemts by setdefault
empty_dict={}
print(empty_dict)
empty_dict.setdefault('key','value')
print(empty_dict)
```

```python
# Create two lists & then merge into a dict
key_values = ['name', 'age', 'rollNo', 'marks']
list_values = ['priiyuuu', 24,66, 90]

# Merge lists into a dictionary
merge_listsAs_dict = dict(zip(key_values, list_values))

# Print the resulting dictionary
print(merge_listsAs_dict)
```

```python
#duplicate items of list and the key is list's items name and value is occurence of ecah itemsin the list
#using 'Counter' class from 'collections' module
from collections import Counter
dulpicate_listItems=[1,3,3,5,6,3,3,4,5,7,7,7]
print("\nlist items are:-",dulpicate_listItems)

dict_ionary1 = dict(Counter(dulpicate_listItems))
print("the merged dictionary will like this:-\n",dict_ionary1)
print("\t")

duplicate_tupleItems=1,2,3,4,5,5,4,3,2,1,2,'q','w','q',
print("\ntuple items are:-",dulpicate_tupleItems)
dict_ionary2=dict(Counter(duplicate_tupleItems))
print("the merged dict like this:-\n",dict_ionary2)
```
