```python
#creating two diffrent list for even and odd
even=[]
odd=[]
list_num=list(range(1,101))
#print(list_num)
for num  in list_num:
    if num%2==0:
        even.append(num)
        #print(num)
    else:
        odd.append(num)
        #print(num)
print("this is even list of numbers:",even,"\n")      
print("this is odd list of numbers:",odd)      
       
```

```python
#creating two list for each even and odd numbers,using concept of comprahension
list_of_numbers=list(range(0,100))
even_No=[n for n in list_of_numbers if n%2==0]
odd_No=[n for n in list_of_numbers if n%2!=0]

print(even_No,"\n")
print(odd_No)
```

```python
#take two lists of same size n make first list members as key and second list elements as values
key_items=['name','age','address']
values_items=['prianka',24,'basaweshwarnagar']
merge_items=dict(zip(key_items,values_items))
print(merge_items)
```

```python
#take one lists of duplicate values,and make a dictionary which key is list element and values is number of occurence  element in that list
from collections import Counter
duplicate_valueslist=[1,2,3,3,4,1,3,4,5,5]
dict_ionary=dict(Counter(duplicate_valueslist))
print(dict_ionary)
```
