```python
# Initialize an empty list to store movie names
movies_names=[]
# infinite loop to collect movie names
while True:
    movie=input("enter movie name(type'STOP' to end):")
    # Check if the user wants to stop (case-insensitive comparison)
    #if  movie=="STOP":     #here if i hvnt mentioned any upper or lower() then it will follow which case i have given LIKE if ihave given ==any digit if you enter that digit thn n thn it will stop otherwise it will your entry as an input
    if movie.upper()=='STOP':  # Fixed: 'stop' changed to 'STOP' to match the uppercase comparison
        # Inform the user that the program recognized the stop command
        print("you said STOP")
        # Exit the loop
        break
    # Add the entered movie name to the list
    movies_names.append(movie)
```

```python
# Calculating sum of numbers from 1 to n (user input)
sum = 0  # Initialize sum to 0
num = int(input("enter the number up to which you want to add:"))  # Get upper limit from user
for i in range(1, num + 1):  # Loop from 1 to num (inclusive)
    sum = sum + i  # Add each number to the running sum
print("The total sum is:", sum)  # Print only the final sum 
```
